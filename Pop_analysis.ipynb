```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Pop.csv")
df.head(3)
```

```python
data = df.iloc[1: , ].reset_index(drop=True)
data
```

```python
data['2000_sum']=  data.iloc[ :,2] + data.iloc[ :, 3]
data['2005_sum']=  data.iloc[ :,4] + data.iloc[ :, 5]
data['2010_sum']=  data.iloc[ :,6] + data.iloc[ :, 7]
data['2015_sum']=  data.iloc[ :,8] + data.iloc[ :, 9]
data['2016_sum']=  data.iloc[ :,10] + data.iloc[ :, 11]
data['2017_sum']=  data.iloc[ :,12] + data.iloc[ :, 13]
data['2018_sum']=  data.iloc[ :,14] + data.iloc[ :, 15]
data['2019_sum']=  data.iloc[ :,16] + data.iloc[ :, 17]
data['2020_sum']=  data.iloc[ :,18] + data.iloc[ :, 19]
```

```python
# 연도별 총 인구

sum_data = data.iloc[:,[0,1,20,21,22,23,24,25,26,27,28]]
sum_data
```

### 1. 연령별 인구분포

```python
age = sum_data.groupby("연령별").sum().reset_index()
age
```

```python
age= age.reindex(index=[0,9,1,2,3,4,5,6,7,8,10,11,12,13,14,15,16,17]).reset_index(drop=True)
age
```

```python
plt.title("연령별 인구분포")
plt.plot(age['2000_sum'],color ="red",label="2000년 연령별 인구분포")
plt.plot(age['2010_sum'],color ="green",label="2010년 연령별 인구분포")
plt.plot(age['2020_sum'],color ="black",label="2020년 연령별 인구분포")
plt.legend()

plt.rc('font', family = 'Malgun Gothic')
plt.xticks(range(18),age["연령별"],rotation=90)
plt.show()
```

### 2. 전국 인구밀집도

```python
city = sum_data.groupby("행정구역별").sum().reset_index()
city
```

```python
city= city.reindex(index=[8,1,10,4,5,6,7,9,0,2,3,11,12,13,14,15]).reset_index(drop=True)
city
```

```python
city2 = city.drop('행정구역별',axis=1)
city2
```

```python
c_area = city2.iloc[0]+city2.iloc[1]+city2.iloc[2]   # 수도권 인구 수 (서울 + 경기 + 인천)
```

```python
other = 0    ## 수도권과 광역시 제외한 나머지 인구수
for i in range(8,16,1):
    other = other + city2.loc[i]
```

```python
total = city2.sum()
```

```python
city2 = city2.append(c_area,ignore_index=True)
city2 = city2.append(other,ignore_index=True)
city2 = city2.append(total,ignore_index=True)
city2
```

```python
city_pop = city2.iloc[[3,4,5,6,7,16,17] , ]  # 수도권 + 광역시 + 그 외 전국 인구 수
city_pop
```

```python
label=(['광주광역시','대구광역시','대전광역시','부산광역시','울산광역시','수도권','그 외 지역'])

plt.title('2000년도 인구 분포')
plt.rc('font', family = 'Malgun Gothic')
plt.pie(city_pop['2000_sum'],labels = label, autopct = '%.1f%%')
plt.show()
```

```python
label=(['광주광역시','대구광역시','대전광역시','부산광역시','울산광역시','수도권','그 외 지역'])
plt.title('2010년도 인구 분포')
plt.rc('font', family = 'Malgun Gothic')
plt.pie(city_pop['2010_sum'],labels = label, autopct = '%.1f%%')
plt.show()
```

```python
label=(['광주광역시','대구광역시','대전광역시','부산광역시','울산광역시','수도권','그 외 지역'])
plt.title('2020년도 인구 분포')
plt.rc('font', family = 'Malgun Gothic')
plt.pie(city_pop['2020_sum'],labels = label, autopct = '%.1f%%')
plt.show()
```
